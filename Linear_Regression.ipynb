```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/drive/MyDrive/AI/car_price_prediction_.csv", nrows=10)
df.columns = df.columns.str.strip()
df.drop(["Car ID"], axis=1, inplace=True)
df
```

```python
plt.scatter(df.Mileage,df.Price)
plt.xlabel("Mileage", color =("red"))
plt.ylabel("Price", color=("red"))
plt.show()
```

```python
lr=linear_model.LinearRegression()
lr.fit(df[['Mileage']],df.Price)
```

```python
#test the model
lr.predict([[100000]])
```

```python
lr.score(df[['Mileage']],df.Price)
lr.coef_
```

```python
lr.intercept_
```
