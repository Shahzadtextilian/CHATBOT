```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv("/content/drive/MyDrive/AI/Experience-Salary.csv")
df.rename(columns={'exp(in months)': 'experience_months', 'salary(in thousands)': 'salary_thousands'}, inplace=True)
display(df.head())
```

```python
plt.scatter(df.experience_months,df.salary_thousands)
plt.xlabel("experience_months", color =("red"))
plt.ylabel("salary_thousands", color=("red"))
plt.show()
```

```python
lr=linear_model.LinearRegression()
lr.fit(df[['experience_months']],df.salary_thousands)
```

```python
#test the model
lr.predict([[100]])
```

```python
lr.score(df[['experience_months']],df.salary_thousands)
lr.coef_
```

```python
lr.intercept_
```
